```python
import re
import nltk
from nltk.tokenize import word_tokenize

# Download tokenizer if not already done
nltk.download('punkt')

def tokenize_vertical(text):
    # 1️⃣ Clean text: remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # 2️⃣ Tokenize
    tokens = word_tokenize(text)
    
    # 3️⃣ Print each token on a new line
    for token in tokens:
        print(token)

# Example text
doc = "Check out this site! It's great."

tokenize_vertical(doc)
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
```

```python
stop_words = list(stop_words)
stop_words.append('https')
stop_words.append('http')
stop_words = set(stop_words)
```

```python
filtered_word = [i for i in wordtoken_doc if not i in stop_words]
print(filtered_word)
```
